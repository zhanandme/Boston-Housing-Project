```python
scipy.stats.levene(boston_df[boston_df["CHAS"]==1]["MEDV"],
                  boston_df[boston_df["CHAS"]==0]["MEDV"])
```

```python
scipy.stats.ttest_ind(boston_df[boston_df["CHAS"]==1]["MEDV"],
                  boston_df[boston_df["CHAS"]==0]["MEDV"])
```

```python
young_medv = boston_df[boston_df['AGE_group'] == 'Young']['MEDV']
middle_medv = boston_df[boston_df['AGE_group'] == 'Middle']['MEDV']
old_medv = boston_df[boston_df['AGE_group'] == 'Old']['MEDV']
```

```python
f_statistic, p_value = scipy.stats.f_oneway(young_medv, middle_medv, old_medv)
print("F_Statistic: {0}, P-Value: {1}".format(f_statistic,p_value))
```
